## Checking the Descriptive Statistics


```python
df[
    ["actual_price", "nett_sales", "nett_profit", "rating_transaction", "rating_branch"]
].describe()
```

**Conclusion:**

## Checking the Data Distribution


```python
columns_to_plot = [
    "actual_price",
    "nett_sales",
    "nett_profit",
    "rating_transaction",
    "rating_branch",
]
columns_to_plot
```

```python
df.hist(figsize=(10, 8), bins=20, column=columns_to_plot, layout=(4, 3))
plt.tight_layout()
```

**Conclusion:**

- The financial variables (actual_price, nett_sales, and nett_profit) show multimodal distributions, indicating the presence of different value segments in the data.
- rating_transaction looks like a continuous variable between 3.0 and 5.0, with many decimal values — meaning users gave a wide variety of ratings.
- rating_branch has fewer rating options, mostly around 4.25–4.75, showing that users tend to give scores in common rounded steps.


## Checking the Data Outliers


```python
df.plot(
    kind="box", figsize=(12, 14), column=columns_to_plot, subplots=True, layout=(4, 4)
)
plt.subplots_adjust(wspace=0.5)
```

**Conclusion:** No Outliers
